# Read in file

```python
import pandas as pd

df = pd.read_csv('GHG_emissions_group.csv')
df.head()
```

# Calculate average emissions per group

```python
avg_emissions = df.groupby('Group', as_index=False)['GHG Emissions per 100g Protein'].mean()
```

# Create the bar graph

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style='whitegrid')

fig, ax = plt.subplots(figsize=(6, 5))
ax = sns.barplot(
    data=avg_emissions,
    x='Group',
    y='GHG Emissions per 100g Protein',
    hue='Group',
    palette = {'Plant': '#A5CCD1', 'Animal': '#F5CDA7'},
    legend=False
)

# Annotate bars with emissions values
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height:.1f}',  # format number to 1 decimal
                (p.get_x() + p.get_width() / 2., height),
                ha='center', va='bottom', fontsize=10)

# Show the plot
plt.title('Average GHG Emissions per 100g Protein by Group')
plt.ylabel('Emissions kg CO2e')
plt.xlabel('Food Group')
plt.tight_layout()
plt.figtext(0.99, 0.01, 'Data collected from Our World in Data (Poore & Nemecek (2018)', ha='right', fontsize=8, color='gray')
plt.savefig('ghg_emissions_chart.png', dpi=300, bbox_inches='tight')
plt.show()

```

# Key takeaways

Plants can provide the same amount of protein with significantly fewer emissions than animal sources.
