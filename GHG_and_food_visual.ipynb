# Read in and display the data

```python
import pandas as pd
file = 'GHG_emissions_protein.csv'

df = pd.read_csv(file)

df.head()
```

```python
# change column names

df = df.rename(columns={
    'GHG emissions per 100g protein (Poore & Nemecek, 2018)': 'GHG Emissions per 100g Protein',
    'Entity': 'Food'
    })

df.sort_values('GHG Emissions per 100g Protein', ascending=False, inplace=True)
df.head()
```

# Group the data

```python
group_map = {
    'Dark Chocolate': 'Plant',
    'Beef (beef herd)': 'Animal',
    'Coffee': 'Plant',
    'Lamb & Mutton': 'Animal',
    'Tomatoes': 'Plant',
    'Prawns (farmed)': 'Animal',
    'Beef (dairy herd)': 'Animal',
    'Berries & Grapes': 'Plant',
    'Cassava': 'Plant',
    'Apples': 'Plant',
    'Cheese': 'Animal',
    'Bananas': 'Plant',
    'Milk': 'Animal',
    'Pig Meat': 'Animal',
    'Citrus Fruit': 'Plant',
    'Rice': 'Plant',
    'Fish (farmed)': 'Animal',
    'Poultry Meat': 'Animal',
    'Brassicas': 'Plant',
    'Root Vegetables': 'Plant',
    'Eggs': 'Animal',
    'Onions & Leeks': 'Plant',
    'Potatoes': 'Plant',
    'Grains': 'Plant',
    'Tofu': 'Plant',
    'Oatmeal': 'Plant',
    'Maize': 'Plant',
    'Wheat & Rye': 'Plant',
    'Groundnuts': 'Plant',
    'Other Pulses': 'Plant',
    'Peas': 'Plant',
    'Nuts': 'Plant'
}

# apply Group to the data frame
df['Group'] = df['Food'].map(group_map)
```

# Create the bar chart

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style='whitegrid')


plt.rcParams['font.family'] = 'Arial'
plt.figure(figsize=(10,8))
sns.barplot(
    data=df,
    x = 'GHG Emissions per 100g Protein',
    y = 'Food',
    hue = 'Group',
    dodge=False,
    palette = {'Plant': '#A5CCD1', 'Animal': '#F5CDA7'}
)

plt.title('Greenhouse Gas Emissions by Food Type', fontsize=14, weight='bold')
plt.xlabel('GHG Emissions per 100g Protein', fontsize=12)
plt.ylabel('Food Type', fontsize=12)
plt.tight_layout()
plt.figtext(0.99, 0.01, 'Data collected from Our World in Data (Poore & Nemecek (2018)', ha='right', fontsize=8, color='gray')
plt.savefig('ghg_emissions_by_food.png', dpi=300, bbox_inches='tight')
plt.show()
```

# Key takeaways

Only plant-based foods appear near the very bottom range, suggesting that the lowest-emission sources of protein are all plant-based.
